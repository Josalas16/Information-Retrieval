```python
import csv

   
#input: dataset1 for the given month    
with open("Flights_dataset1_July.txt", "r") as csv_file: 
    csv_reader = csv.reader(csv_file, delimiter=',')
    
    elp_flights = []
    for row in csv_reader:
        x = row[2].find("El Paso") #to check for outgoing (row[1]) or incoming (row[2]) flights  
        if x != -1 and row[3] != 1: #to check if the current row have info about ELP flights AND is not cancelled
            row[0] = row[0].replace("-","") #to change format of the date to YYYYMMDD
            elp_flights.append(row)
            print(row)
    

 #change file name to desire month    
f = open("elp_incoming_flights_datasaset1_july.csv", "w")

for row in elp_flights:
    print(type(row))
    f.write("%s\n" % ", ".join(row))
    
f.close()




 
```
