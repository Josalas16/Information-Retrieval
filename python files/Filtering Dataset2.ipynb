```python
import csv

   
#input: dataset1 for the given month    
with open("Flights_dataset2_cleaned_July.txt", "r") as csv_file: 
    csv_reader = csv.reader(csv_file, delimiter=',')
    
    elp_flights = []
    for row in csv_reader:
        x = row[2].find("ELP") #to check for outgoing (row[1]) or incoming (row[2]) flights  
        if x != -1: #to check if the current row have info about ELP flights
            elp_flights.append(row)
            print(row)
    

 #change file name to desire month and incoming/outgoing    
f = open("elp_incoming_flights_datasaset2_july.csv", "w")

for row in elp_flights:
    print(type(row))
    f.write("%s\n" % ", ".join(row))
    
f.close()
```
